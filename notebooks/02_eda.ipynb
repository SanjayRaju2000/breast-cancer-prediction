# Exploratory Data Analysis

### Notebook Summary 

This notebook explores the dataset to uncover patterns, relationships, and insights that will inform the preprocessing and modeling stages. The analysis includes:  

- Examining the overall structure and distributions of numerical and categorical features.  
- Identifying potential outliers and unusual values in continuous variables such as age, BMI, and cholesterol level.  
- Exploring the balance of categorical features such as gender, cancer stage, smoking status, and treatment type.  
- Assessing the distribution of the target variable (`survived`) to understand class balance.  
- Investigating relationships between independent variables and the survival outcome, using statistical summaries and visualizations.  

The goal of this notebook is to build an understanding of the dataset, highlight important predictive features, and identify any potential challenges (such as class imbalance or skewed distributions) that may affect model performance.  


### Notebook Setup

```python
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Loading

```python
# Loading dataset
cancer_df = pd.read_csv('D:/lc-mortality/data/Lung Cancer.csv')

# View dataset
cancer_df.head(10)
```

### Distribution of Target Variable

```python
cancer_df['survived'].value_counts()
```

```python
survival_counts = cancer_df['survived'].value_counts()

plt.bar(survival_counts.index, survival_counts.values, color=['salmon','skyblue'])
plt.xticks([0, 1], ['Died', 'Survived'])
plt.ylabel('Number of Patients')
plt.title('Distribution of Survival Outcomes')
plt.show()
```

### Smoking Status

```python
# Smoking status bar chart

cancer_df['smoking_status'].value_counts()
```

```python
smoking_status_counts = cancer_df['smoking_status'].value_counts()

sns.countplot(data=cancer_df, x='smoking_status', hue='survived', palette='Set2')
plt.ylabel('Number of Patients')
plt.xlabel('Smoking Status')
plt.title('Survival Outcome by Smoking Status')
plt.xticks(rotation=20)
plt.show()
```

### Gender

```python
sns.countplot(data=cancer_df, x='gender', hue='survived', palette='Set2')
plt.ylabel('Number of Patients')
plt.xlabel('Gender')
plt.title('Survival Outcome by Gender')
plt.xticks(rotation=20)
plt.show()
```

### Treatment type

```python
cancer_df['treatment_type'].value_counts()
```

```python
treatment_type_counts = cancer_df['treatment_type'].value_counts()

sns.countplot(data=cancer_df, x='treatment_type',hue='survived', palette='Set2')
plt.ylabel('Number of Patients')
plt.xlabel('Treatment type')
plt.title('Survival Outcome by Treatment type')
plt.xticks(rotation=20)
plt.show()
```

### Cancer Stage

```python
cancer_df['cancer_stage'].value_counts()
```

```python
stage_counts = cancer_df['cancer_stage'].value_counts()

sns.countplot(data=cancer_df, x='cancer_stage', hue='survived', palette='Set2')
plt.ylabel('Number of Patients')
plt.xlabel('Cancer Stage')
plt.title('Survival by Cancer Stage')
plt.show()
```

### BMI

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=cancer_df, x='survived', y='bmi', hue='survived', palette='Set2', dodge=False)
plt.xticks([0,1], ['Did Not Survive (0)', 'Survived (1)'])
plt.ylabel('BMI')
plt.title('BMI Distribution by Survival Outcome')
plt.show()
```
